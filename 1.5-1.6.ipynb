```python
# посчитаем среднее значение math_score, reading_score в зависимости от пола студентов
students_performance.groupby('gender').aggregate({'math_score': 'mean', 'reading_score': 'mean'})
# aggregate принимает слооварь, содержащий переменную и что по этой переменной мы хотим узнать

# Результатом агрегации будет датафрейм, с индексами из тех значений, по которым мы проводили агрегацию
# Если укажем агрумент as_index=False, то индексами будут цифры от 0 до n

students_performance.groupby('gender', as_index=False).aggregate({'math_score': 'mean', 'reading_score': 'mean'}).rename(columns = {
    'math_score': 'mean_math_score', 'reading_score': 'mean_reading_score'
})
```

```python
# Можем группировать по нескольким переменным. Для этого оборачиваем переменные для группировки в список.
# Уберем индексы и получим сложный составной индекс
mean_scores = students_performance.groupby(['gender', 'race/ethnicity']).aggregate({'math_score': 'mean', 'reading_score': 'mean'}).rename(columns = {
    'math_score': 'mean_math_score', 'reading_score': 'mean_reading_score'
})
```
