```python
# сообщаем Jupyter Notebook'у, что мы хотим получать визуализации прямо тут
%matplotlib inline 
import matplotlib.pyplot as plt
import seaborn as sns # надстройка на mathplotlib с более простым синтаксисом. Более красивые графики
```

```python
students_performance = pd.read_csv('https://stepik.org/media/attachments/course/4852/StudentsPerformance.csv')
```

```python
students_performance
```

```python
students_performance['math score'].hist() # вывести гистограмму
```

```python
students_performance.plot.scatter(x='math score', y='reading score') # получим скаттер-плот
```

```python
sns.lmplot(x='math score', y='reading score', data=students_performance) 
# показывает облако точек и регрессионную прямую, которая показывает линейную апроксимацию взаимосвязи
```

```python
# добавим группирующую переменную
sns.lmplot(x='math score', y='reading score', hue = 'gender', data=students_performance) 
# агрумент fit_reg=False убирает регрессионные прямые
```

```python
ax = sns.lmplot(x='math score', y='reading score', hue = 'gender', fit_reg=False, data=students_performance) 
ax.set_xlabels('Math score') # Подписываем ось Х
ax.set_ylabels('Reading score') # Подписываем ось У
```
