```python
my_stat = pd.read_csv('https://stepik.org/media/attachments/course/4852/my_stat.csv')
my_stat
```

```python
# В переменную с именем subset_1 сохраните только первые 10 строк и только 1 и 3 колонку (номера, а не индексы).

subset_1 = my_stat.iloc[0:10, [0, 2]]  # если срез, то без квадратных скобок. без среза - [[],[]]
subset_1
```

```python
subset_2 = my_stat.drop([0,4]) # удалим 1 и 5 строки (номера, а не индексы) axis=0 - строки (по умолчанию 0)
subset_2 = subset_2.iloc[:,[1, 3]]
subset_2
```

```python
# В переменную subset_1 сохраните только те наблюдения, у которых значения переменной V1  строго больше 0, 
# и значение переменной V3  равняется 'A'.

subset_1 = my_stat.query('V1 > 0 & V3 == "A"')
subset_1
```

```python
# В переменную  subset_2  сохраните только те наблюдения, у которых значения переменной V2  не равняются 10, 
# или значения переменной V4 больше или равно 1.
subset_2 = my_stat.query('V2 != 10 | V4 >= 1')
subset_2
```

```python
# Теперь давайте преобразуем наши данные. В переменной my_stat лежат данные с которыми вам необходимо проделать 
#следующее дейтвие. В этих данных (my_stat) создайте две новые переменных:

# V5 = V1 + V4
# V6 = натуральный логарифм переменной V2

my_stat['V5'] = my_stat['V1'] + my_stat['V4']
my_stat['V6'] = np.log(my_stat['V2'])
my_stat
```

```python
# Переименуйте колонки в данных  my_stat следующим образом:
# V1 -> session_value
# V2 -> group
# V3 -> time
# V4 -> n_users

my_stat = my_stat.rename(columns={'V1':'session_value', 'V2':'group', 'V3':'time', 'V4': 'n_users'})
my_stat
```

```python
my_stat = pd.read_csv('https://stepik.org/media/attachments/course/4852/my_stat_1.csv')

# В переменной session_value замените все пропущенные значения на нули.

my_stat['session_value'] = my_stat['session_value'].fillna(0)
my_stat
```

```python
# В переменной n_users замените все отрицательные значения на медианное значение переменной n_users (без учета
# отрицательных значений, разумеется).

median_n_users = my_stat.query("n_users >= 0")['n_users'].median() # найдем медиану неотрицательных
my_stat.loc[my_stat['n_users'] < 0, 'n_users'] = median_n_users # заменяем отрицательные значения на медиану
my_stat
```

```python
# Обратите внимание, что при такой записи группирующие переменные станут индексами в итоговом dataframe

# Также функция count() применится ко всем колонкам, что не всегда является желанным результатом. Чтобы применить функцию 
# только к нужной колонке в данных, можно использовать связку  groupby() + agg()
```

```python
# В этой задаче для данных my_stat рассчитайте среднее значение переменной session_value для каждой группы (переменная group),
# в получившемся dataframe  переменная group не должна превратиться в индекс. Также переименуйте колонку со средним 
# значением session_value в mean_session_value.

# Получившийся результат сохраните в dataframe с именем mean_session_value_data.

my_stat = pd.read_csv('https://stepik.org/media/attachments/course/4852/my_stat_1.csv')
my_stat
```

```python
mean_session_value_data = my_stat.groupby('group', as_index=False).aggregate({'session_value': 'mean'}) 
# С аргументом as_index=False - группирует, не вынося в индекс (иначе значения 'group' служили бы индексами)
mean_session_value_data = mean_session_value_data.rename(columns={'session_value':'mean_session_value'})
mean_session_value_data
```
